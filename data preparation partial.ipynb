```python
def data_create(file_path ,start_row ):
    start_row = start_row 
    num_rows_to_read = 1000000 
    chunk_size = 10000
    result_df = pd.DataFrame()
    num_chunks = num_rows_to_read // chunk_size

    for i in range(num_chunks):
        skip_rows = start_row + i * chunk_size
        chunk_df = pd.read_csv(file_path, skiprows=range(1, skip_rows), nrows=chunk_size, header=None)
        result_df = pd.concat([result_df, chunk_df], ignore_index=True)
        df = result_df
        data = df.drop(df.columns[10:] , axis=1)
        data.rename(columns = {0: 'token_number' , 1: 'timestamp' , 2: 'time_entry' , 3: 'volume' , 4 : 'open' , 5: 'high' , 6:'low' , 7:'prev-close' , 8:'ltp' ,9:'atp'} , inplace=True)
        print(data)
        data['token_number'] = data['token_number'].apply(lambda x: x.split(':')[-1])
        data['token_number']=data['token_number'].apply(lambda x: x[2:].strip(" '"))
        data['timestamp'] = data['timestamp'].apply(lambda x: x.split('p')[1])
        data['timestamp'] = data['timestamp'].apply(lambda x: x[2:].strip(" '"))
        data['time_entry'] = data['time_entry'].apply(lambda x: x.split('y')[-1])
        data['time_entry'] = data['time_entry'].apply(lambda x: x[2:].strip(" '"))
        data['volume'] = data['volume'].apply(lambda x: x.split(':')[-1])
        data['open'] = data['open'].apply(lambda x: x.split(':')[-1])
        data['high'] = data['high'].apply(lambda x: x.split(':')[-1])
        data['low'] = data['low'].apply(lambda x: x.split(':')[-1])
        data['prev-close'] = data['prev-close'].apply(lambda x: x.split(':')[-1])
        data['ltp'] = data['ltp'].apply(lambda x: x.split(':')[-1])
        data['atp'] = data['atp'].apply(lambda x: x.split(':')[-1])
        
        data.to_csv('option_data_file' , index = False)
       
```
